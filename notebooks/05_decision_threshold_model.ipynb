# 05 - Decision Threshold Model
Apply LiquidityReadinessAssessor and expose an interactive widget for operational decision making.

```python
import sys
import os
import ipywidgets as widgets
from IPython.display import display

# Add parent dir to sys.path to import framework
sys.path.append(os.path.abspath('..'))

from framework.assessor import LiquidityReadinessAssessor
```

```python
assessor = LiquidityReadinessAssessor()
spread = widgets.FloatSlider(description='TWAS %', value=1.0, min=0.1, max=5.0, step=0.1)
depth = widgets.IntSlider(description='Depth 1% USD', value=100000, min=10000, max=1000000, step=10000)
quote = widgets.FloatSlider(description='Quote %', value=95.0, min=70, max=100, step=0.1)
resilience = widgets.FloatSlider(description='Recovery min', value=5.0, min=0.5, max=30, step=0.5)
turnover = widgets.FloatSlider(description='Vol/Supply %', value=0.2, min=0.01, max=2.0, step=0.01)
out = widgets.Output()

def on_change(_=None):
    with out:
        out.clear_output()
        result = assessor.assess({
            'time_weighted_avg_spread_pct': spread.value,
            'avg_depth_1.0pct_notional': depth.value, # Maps to framework metric key
            'quote_presence_ratio_pct': quote.value,
            'spread_resilience_min': resilience.value,
            'daily_volume_to_circulating_supply_pct': turnover.value,
        })
        print(f"RECOMMENDED MODE: {result.mode}")
        print("\nReasoning:")
        for r in result.reasons:
            print('-', r)

for w in [spread, depth, quote, resilience, turnover]:
    w.observe(on_change, names='value')

print("Adjust sliders to see liquidity readiness recommendation:\n")
display(spread, depth, quote, resilience, turnover, out)
on_change()
```
